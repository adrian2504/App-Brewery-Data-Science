### Gradient descent

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1 
precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(700):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print('loop ran this many times', n)
        break 
        
print('Local minimum occurs at:', new_x)
print('slope or valur of df(x) value at this point is: ', df(new_x))
print('f(x) value or cost at this point is:',f(new_x))
```

```python
# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[20,5])

# 1 Chart: cost function

plt.subplot(1,3,1)
plt.title('Cost function',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1, f(x_1),color='blue',linewidth=3, alpha=0.6)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='r', s=100) # a function can work with an array but not with a list

# 2 Chart: Derivative

plt.subplot(1,3,2)
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.grid()
plt.title('Slope of the cost function',fontsize=17)
plt.plot(x_1, df(x_1), color='red', linewidth=3, alpha=0.6)


plt.scatter(x_list,slope_list,color='y',s=100 ,alpha=0.6)


# 3 Chart: Derivative (Close up)

plt.subplot(1,3,3)
plt.title('Gradient Descent (close up)',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.grid()
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)


plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)


plt.scatter(x_list,slope_list,color='y',s=100 ,alpha=0.6)

plt.show()
```

## Gradient descent as python algorithm

```python

def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iteration=300): 
    
    new_x = initial_guess 
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iteration):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x) 
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return ( new_x, x_list, slope_list)
        
```

##  Minimise $$ f(x,y) = \frac{1}{3^{-x^2 - y^2}+1}$$

Minimise $$f(x,y) = \frac{1}{r+1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
def f(x,y):
    r = 3**(-x**2 - y**2) 
    return( 1/(r+1))
```

## Partial Derivatives and Symbolic Computation

## $$ \frac{ \partial f}{ \partial x} = \frac{ 2x \ln(3) \cdot 3^{-x^2 - y^2}}{\left (3^{-x^2-y^2}+1 \right) ^2} $$


## $$ \frac{ \partial f}{ \partial y} = \frac{ 2y \ln(3) \cdot 3^{-x^2 - y^2}}{\left (3^{-x^2-y^2}+1 \right) ^2} $$

```python
a, b = symbols('x,y')
print('Our cost function f(x,y): ',f(a,b))

diff_x = diff(f(a,b),a)
print('Partial derivative of f(x,y) wrt x:',diff_x)

# evaluating my function
print('Value of f(x,y) at x=1.8 and y=1.0 is:',
      f(a,b).evalf(subs={a:1.8,b:1.0}) )
      # the curly braces within indicates indicates python dictionaries
    
# evaluating derivative of the function
print('Value of df(x,y) at x=1.8 and y=1.0 is:',
      diff_x.evalf(subs={a:1.8,b:1.0}) )
      # the curly braces within indicates indicates python dictionaries
```

## Batch Gradient Descent with SymPy

```python
# Setup
multiplier = 0.1
max_iter = 200
params = np.array([1.8,1.0]) #initial guess

for n in range(max_iter):
    gradient_x = diff(f(a,b),a).evalf(subs={a:params[0], b:params[1]})
    gradient_y = diff(f(a,b),b).evalf(subs={a:params[0], b:params[1]})
    
    gradients = np.array([gradient_x, gradient_y ])
    
    params = params - multiplier * gradients

print('Values in gradient array', gradients)
print('Minimum occurs at x value of : ',params[0])
print('Minimum occurs at y value of : ',params[1])
print('The cost is ',f(params[0],params[1]))

```

```python
# partial derivative functions example 4

from math import log as log

def fpx(x,y):
    r = 3**(-x**2 - y**2)
    return 2*x*log(3)*r/(r+1) **2

def fpy(x,y):
    r = 3**(-x**2 - y**2)
    return 2*x*log(3)*r/(r+1) **2
```

## Example 5 - working with data and a real Cost Function
### Mean Squared Error: A cost function for regression problems

#### $$ RSS = \sum_{i=1} ^{n} \big(y^{(i)} - h_\theta x ^ {(i)} \big) ^2$$

#### $$ MSE = \frac {1}{n} \sum_{i=1} ^{n} \big(y^{(i)} - h_\theta x ^ {(i)} \big) ^2$$

#### $$ MSE = \frac {1}{n} \sum_{i=1} ^{n} \big(y - \hat{y} \big) ^2$$

```python
# make sample data 

x_5 = np.array([0.1,1.2,2.4,3.2,4.1,5.7,6.5])
y_5 = np.array([1.7,2.4,3.5,3.0,6.1,9.4,8.2])


x_5 = x_5.reshape(len(x_5),1)
y_5 = y_5.reshape(len(y_5),1)

print('Shape of x_5 array: ',x_5.shape)
print('Shape of y_5 array: ',y_5.shape)



```

```python
# Quick linear regression

# linear regression requires a 2D numpy array 
regr = LinearRegression()
regr.fit(x_5,y_5)

print('Theta 0: ',regr.intercept_[0])
print('Theta 1:', regr.coef_[0][0])
```

```python
plt.scatter(x_5,y_5,s=50)
plt.plot(x_5,regr.predict(x_5), color='orange', linewidth=3)
plt.xlabel('X label')
plt.ylabel('Y label')

plt.show()
```

```python
# y_hat = theta0 + theta1*x

y_hat = 0.8475351486029536 + 1.2227264637835915 *x_5
print('Est values y_hat are: \n',y_hat)
print('In comparison the actual value y values are \n',y_5)
```

```python

def mse(y, y_hat):
    #mse_cal = 1/7*sum((y-y_hat)**2)
    #mse_cal = (1/y.size)*sum((y-y_hat)**2)
    mse_cal = np.average(y - y_hat**2, axis = 0) # mentioning axis=0 averages rows
    return mse_cal
```
